```python
# importing the libraries
import pandas as pd
import numpy as np
import requests
```

```python

# Scraping the table
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
postcodes = pd.read_html(url.content, header=0)[0]
```

```python
df = postcodes[postcodes.Neighbourhood != 'Not assigned']
df.head()
```

```python
df.groupby(['Postal Code']).first()
df[df['Borough'] == 'Not assigned']
```

```python
url2 = 'http://cocl.us/Geospatial_data'
df_coor = pd.read_csv(url2)
df_coor.head()
```

```python
dataframe = df.join(df_coor.set_index('Postal Code'), on='Postal Code')
dataframe
```
